```python
%load_ext autoreload
%autoreload 2

import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output, display

env = gym.make('Taxi-v3')
env.reset()
print (env.action_space)
print (env.observation_space)
next_state, reward, terminated, truncated, info = env.step(1)
print ()
print ('next_state',next_state)
print ('reward',reward)
print ('terminated',terminated)
print ('truncated',truncated)
print ('info',info)
```

```python
episodes = 5000
test_episodes = 100
gamma = 0.99
alpha = 0.1
epsilon = 0.2

Q = np.zeros((env.observation_space.n, env.action_space.n))
print (Q.shape)
```

```python
def epsilon_greedy(Q, state, n_actions, epsilon=epsilon):
    if np.random.rand() < epsilon:
        return np.random.randint(n_actions)
    else:
        return np.argmax(Q[state])

"""Sarsa 算法， 更新Qtable"""
def train_sarsa(env, episodes, alpha, gamma):
    Q = np.zeros((env.observation_space.n, env.action_space.n))
    rewards = []

    for ep in range(episodes):
        state, _ = env.reset()
        action = epsilon_greedy(Q, state, env.action_space.n)
        total_reward = 0
        done = False

        while not done:
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            # choose next action
            next_action = epsilon_greedy(Q, next_state, env.action_space.n)

            # Sarsa : update Q table
            Q[state, action] += alpha * (reward + gamma * Q[next_state, next_action] - Q[state, action])
            
            state, action = next_state, next_action
            total_reward += reward

        rewards.append(total_reward) # 一整个episode的rewards

    return Q, rewards


""" Qlearning """
def train_Qlearning(env, episodes, alpha, gamma):
    Q = np.zeros((env.observation_space.n, env.action_space.n))
    rewards = []

    for ep in range(episodes):
        state, _ = env.reset()
        action = epsilon_greedy(Q, state, env.action_space.n)
        total_reward = 0
        done = False

        while not done:
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            # choose next action
            next_action = epsilon_greedy(Q, next_state, env.action_space.n)

            # Sarsa : update Q table 
            # The only difference with Sarsa: np.max(Q[next_state])
            Q[state, action] += alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])
            
            state, action = next_state, next_action
            total_reward += reward

        rewards.append(total_reward) # 一整个episode的rewards

    return Q, rewards
```

```python
""" 测试训练得到的Q table 的效果，
    返回每次episode的奖励的均值"""
def test_policy(env, Q, episodes=100):
    total_rewards=[]

    for ep in range(episodes):
        state, _ = env.reset()
        done = False
        total_reward = 0

        while not done:
            action = np.argmax(Q[state])
            state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            total_reward += reward

        total_rewards.append(total_reward)
        
    return np.mean(total_rewards)


def run_demo_notebook(env, Q, episodes=1, delay=0.5):
    """
    在 Jupyter Notebook 中可视化 Taxi-v3 策略执行
    """
    for ep in range(episodes):
        state, _ = env.reset()
        done = False
        total_reward = 0
        print(f"\n=== 演示 Episode {ep+1} ===")

        while not done:
            # 获取 ASCII 渲染
            frame = env.render()
            
            # 在 Notebook 中显示
            clear_output(wait=True)
            print(frame)
            
            # 按贪婪策略选择动作
            action = np.argmax(Q[state])
            state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            total_reward += reward
            
            time.sleep(delay)
        
        # 最后一次显示
        clear_output(wait=True)
        print(env.render())
        print(f"Episode {ep+1} 总奖励: {total_reward}")
```

```python

if __name__ == "__main__":
    env = gym.make("Taxi-v3")

    # 训练 Sarsa
    Q_table_sarsa, rewards_sarsa = train_sarsa(env, episodes, alpha, gamma)
    # print (Q_table_sarsa.shape)
    # print (Q_table_sarsa[:5])
    mean_reward_sarsa = test_policy(env, Q_table_sarsa)

    # 训练 Qlearning
    Q_table_Qlearning, rewards_Qlearning = train_Qlearning(env, episodes, alpha, gamma)
    mean_reward_Qlearning = test_policy(env, Q_table_Qlearning)

    # 用 convolce 做滑动平均， 滑动窗口为100
    # 因为reward随机性很大
    plt.plot(np.convolve(rewards_sarsa, np.ones(100)/100, mode='valid'), label="Sarsa")
    plt.plot(np.convolve(rewards_Qlearning, np.ones(100)/100, mode='valid'), label="Qlearning")
    plt.xlabel("Episodes")
    plt.ylabel("Average Reward (window=100)")
    plt.title("Taxi-v3: Sarsa vs Q-learning")
    plt.legend()
    plt.show()
```

### 问题：关注你使用的算法中的细节（例如动作的探索（Exploration）和利用（Exploitation）、经验回放（Experience Replay）），分析它们在该算法中被使用的原因

**探索和利用：**
1.上面的算法中用了epsilon-greedy来平衡探索和利用。
2.对于sarsa和qlearning的值迭代，采取了不同的策略。

- sarsa: 是on-policy, 更新时依赖下一个实际执行的动作a'，因此他收敛更稳定，但是收敛速度更慢。
  由此看出，sarsa更保守
- qlearning： 是off-policy，更新时选取最大q值的动作，收敛速度更快（上面图片中，橙色的线高于蓝色的线），但是收敛可能震荡，尤其当探索不足是。
  由此看出，qlearning更为激进，而却需要更多的“利用（exploitation）”

**经验回放：**
Sarsa和Qlearning中没有用经验回放。在算法中，我们采用边采样边更新的方法（online learning），因此没有用经验回放。
但是在DQN中，用神经网络近似Q值，直接每次更新相当于随机梯度下降，梯度更新会震荡严重，因此会采用经验回放的方式：存储（s,a,r,s'）四元组， 让学习过程更稳定。

### 问题：了解Reward Shaping技巧，尝试在训练过程中使用它，并记录它带来的影响

**答：**
Reward Shaping 的想法是：给智能体额外的中间奖励（辅助信号），引导它更快学到有价值的策略。
一个简单的应用是用曼哈顿距离（Manhattan Distance）: F = -0.1 * (距离目的地)

```python
env = gym.make("Taxi-v3", render_mode="ansi")
env.reset()
next_state, reward, terminated, truncated, _ = env.step(0) 
print (next_state)

taxi_row, taxi_col, pass_loc, dest = env.unwrapped.decode(next_state)
dest_row, dest_col = [(0,0),(0,4),(4,0),(4,3)][dest]
print ('taxi_row', taxi_row)
print ('taxi_col', taxi_col)
print ('pass_loc', pass_loc)
print ('dest', dest)
print ('', )

### 给 state 和 reward， 返回 shaping 
def shaping_reward(state, env):
    taxi_row, taxi_col, pass_loc, dest = env.unwrapped.decode(state)
    dest_row, dest_col = [(0,0),(0,4),(4,0),(4,3)][dest]
    
    dist = abs(taxi_row - dest_row) + abs(taxi_col - dest_col)
    shaping = -0.1 * dist
    return shaping
```

```python
"""Qlearning with reward shaping"""
def train_Qlearning_rshaping(env, episodes, alpha, gamma):
    Q = np.zeros((env.observation_space.n, env.action_space.n))
    rewards = []

    for ep in range(episodes):
        state, _ = env.reset()
        action = epsilon_greedy(Q, state, env.action_space.n)
        total_reward = 0
        done = False

        while not done:
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            ### Here, add shaping reward
            reward += shaping_reward(state, env)

            # choose next action
            next_action = epsilon_greedy(Q, next_state, env.action_space.n)

            # Sarsa : update Q table 
            # The only difference with Sarsa: np.max(Q[next_state])
            Q[state, action] += alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])
            
            state, action = next_state, next_action
            total_reward += reward

        rewards.append(total_reward) # 一整个episode的rewards

    return Q, rewards
```

```python

env = gym.make("Taxi-v3")

# 训练 Sarsa
Q_table_sarsa, rewards_sarsa = train_sarsa(env, episodes, alpha, gamma)
mean_reward_sarsa = test_policy(env, Q_table_sarsa)

# 训练 Qlearning
Q_table_Qlearning, rewards_Qlearning = train_Qlearning(env, episodes, alpha, gamma)
mean_reward_Qlearning = test_policy(env, Q_table_Qlearning)

# 训练 Qlearning with reward shaping
Q_table_Qlearning_rshaping, rewards_Qlearning_rshaping = train_Qlearning_rshaping(env, episodes, alpha, gamma)
mean_reward_Qlearning_rshaping = test_policy(env, Q_table_Qlearning_rshaping)

# 用 convolce 做滑动平均， 滑动窗口为100
# 因为reward随机性很大
plt.plot(np.convolve(rewards_sarsa, np.ones(100)/100, mode='valid'), label="Sarsa")
plt.plot(np.convolve(rewards_Qlearning, np.ones(100)/100, mode='valid'), label="Qlearning")
plt.plot(np.convolve(rewards_Qlearning_rshaping, np.ones(100)/100, mode='valid'), label="Qlearning_rshaping")
plt.xlabel("Episodes")
plt.ylabel("Average Reward (window=100)")
plt.title("Taxi-v3: Sarsa vs Q-learning")
plt.legend()
plt.show()
```

**但是发现加了 reward shaping 后并没有提升， 反而比 Qlearning 的曲线要低一点
推断可能是因为环境太简单了，并且有因为存在障碍物的原因，有时候出租车需要先远离dest，
然后再绕过障碍物去dest，这样rewar会减少一些，因此收敛速度会稍微慢一些。**
